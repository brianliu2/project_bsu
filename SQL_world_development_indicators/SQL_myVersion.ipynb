#### JOIN: OUTER JOIN, INNER JOIN and LEFT JOIN

1. OUTER JOIN: Merge two tables without considering any matching criterion.
2. INNER JOIN: Merge two tables with having entries from both tables that satisfies matching criterion.
3. LEFT  JOIN: Merge two tables with having all entries from left table regardless matching criterion, but only the matching entries from right table.


As being an example, we only consider this country table in this database without involving other tables for this illustrative purpose, but perform subqueries to 'create' two sub-tables by only selecting five countries with a few columns from original table; sub-table A: table for countries (AFG, ALB, ASM, BEL) and sub-table B: table for countries (AFG, ARM, URY, BEL)

INNER JOIN followed by two subqueries

```python
pd.read_sql(
        """ 
           
            SELECT       A.CountryCode AS CountryCode
                        ,A.LatestPopulationCensus AS LatestPopulationCensus
                        ,B.SourceOfMostRecentIncomeAndExpenditureData AS SourceOfMostRecentIncomeAndExpenditureData
                        ,A.ShortName AS ShortName
            FROM       ( 
                            
                            
                           SELECT      CountryCode
                                        ,LatestPopulationCensus
                                        ,SourceOfMostRecentIncomeAndExpenditureData
                                        ,ShortName
                           FROM        Country
                           WHERE       CountryCode IN ('AFG','ALB', 'ASM', 'BEL')
                        ) AS A
            INNER JOIN   (
                            
                            
                            SELECT      CountryCode
                                        ,LatestPopulationCensus
                                        ,SourceOfMostRecentIncomeAndExpenditureData
                                        ,ShortName
                            FROM        Country 
                            WHERE       CountryCode IN ('AFG','ARM', 'URY', 'BEL')
                            
                          ) AS B
            ON          A.CountryCode = B.CountryCode    
            
        """, con=conn)
```

LEFT JOIN followed by two subqueries

```python
pd.read_sql(
        """ 
           
            SELECT       A.CountryCode AS CountryCode
                        ,A.LatestPopulationCensus AS LatestPopulationCensus
                        ,B.SourceOfMostRecentIncomeAndExpenditureData AS SourceOfMostRecentIncomeAndExpenditureData
                        ,A.ShortName AS ShortName
            FROM       ( 
            
                            
                           SELECT      CountryCode
                                        ,LatestPopulationCensus
                                        ,SourceOfMostRecentIncomeAndExpenditureData
                                        ,ShortName
                           FROM        Country
                           WHERE       CountryCode IN ('AFG','ALB', 'ASM', 'BEL')
                        ) AS A
            LEFT JOIN   (
            
                            
                            SELECT      CountryCode
                                        ,LatestPopulationCensus
                                        ,SourceOfMostRecentIncomeAndExpenditureData
                                        ,ShortName
                            FROM        Country
                            WHERE       CountryCode IN ('AFG','ARM', 'URY', 'BEL')
                            
                          ) AS B
            ON          A.CountryCode = B.CountryCode    
            
        """, con=conn)
```

#### UNION, INTERSECT and EXCEPT**

1. UNION: combines entries from two or more SELECT - FROM statements. PS: UNION removes duplicated entries; UNION ALL includes all entries duplicates.

2. INTERSECT: output entries existing both queries.

3. EXCEPT: output entries only existing the top query but not the bottom one.

```python
# UNION 
pd.read_sql(
        """ 
                           SELECT      CountryCode
                                        ,LatestPopulationCensus
                                        ,SourceOfMostRecentIncomeAndExpenditureData
                                        ,ShortName
                           FROM        Country
                           WHERE       CountryCode IN ('AFG','ALB', 'ASM', 'BEL')
                       
                           UNION
                           
                           SELECT      CountryCode
                                        ,LatestPopulationCensus
                                        ,SourceOfMostRecentIncomeAndExpenditureData
                                        ,ShortName
                           FROM        Country AS A
                           WHERE       CountryCode IN ('AFG','ARM', 'URY', 'BEL')
            
        """, con=conn)
```

```python
# INTERSECT 
pd.read_sql(
        """ 
                           SELECT      CountryCode
                                        ,LatestPopulationCensus
                                        ,SourceOfMostRecentIncomeAndExpenditureData
                                        ,ShortName
                           FROM        Country
                           WHERE       CountryCode IN ('AFG','ALB', 'ASM', 'BEL')
                       
                           INTERSECT
                           
                           SELECT      CountryCode
                                        ,LatestPopulationCensus
                                        ,SourceOfMostRecentIncomeAndExpenditureData
                                        ,ShortName
                           FROM        Country AS A
                           WHERE       CountryCode IN ('AFG','ARM', 'URY', 'BEL')
            
        """, con=conn)
```

```python
# EXCEPT 
pd.read_sql(
        """ 
                           SELECT      CountryCode
                                        ,LatestPopulationCensus
                                        ,SourceOfMostRecentIncomeAndExpenditureData
                                        ,ShortName
                           FROM        Country
                           WHERE       CountryCode IN ('AFG','ALB', 'ASM', 'BEL')
                       
                           EXCEPT
                           
                           SELECT      CountryCode
                                        ,LatestPopulationCensus
                                        ,SourceOfMostRecentIncomeAndExpenditureData
                                        ,ShortName
                           FROM        Country AS A
                           WHERE       CountryCode IN ('AFG','ARM', 'URY', 'BEL')
            
        """, con=conn)
```
